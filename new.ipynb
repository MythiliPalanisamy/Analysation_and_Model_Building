```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
final1 = pd.read_csv('final1.csv')
```

correlation of 'price' with other variable

```python
columns_to_include = ['Price', 'Bedrooms', 'Primary energy consumption', 'Furnished', 'Terrace', 'Terrace surface', 'Surface of the plot', 'Living room surface', 'Number of frontages', 'Construction year', 'Office', 'Toilets', 'Shower rooms']
sns.pairplot(data=final1, vars=columns_to_include)
```

```python
numeric_columns = final1.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()
correlation_matrix
```

```python
sns.scatterplot( x='Price',y= 'Shower rooms', data=final1, size='Shower rooms', sizes=(100,500))
```

Represent the number of properties according to their surface using a histogram.

```python
sns.set_style('whitegrid')
dark_palette = sns.color_palette('dark', n_colors=11)
sns.histplot(x='Surface of the plot', data=final1, binrange=(1,20000), binwidth=500, hue='province',multiple='stack' , palette=dark_palette)
```
